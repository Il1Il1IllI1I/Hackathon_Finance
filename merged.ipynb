```python
import pandas as pd

# 각 sector에 대한 이름 리스트
name_list = ['construction', 'discretionary', 'energy', 'finance', 'heavy', 'IT', 'staples', 'steels']

# 각 이름에 대해 루프를 돌며 데이터를 병합하고 저장
for name in name_list:
    # 기술 지표 데이터 로드 (data1)
    data1 = pd.read_csv(f'/mnt/data/{name[:3]}fin.csv', index_col='Date', parse_dates=True)
    
    # 기준이 될 데이터 로드 (data2)
    data2 = pd.read_csv(f'real_dataset/{name}.csv', index_col='날짜', parse_dates=True)

    # '날짜'를 기준으로 inner join
    merged_data = pd.merge(data1, data2, left_index=True, right_index=True, how='inner')

    # 결과 확인
    print(f'--- Merged Data for {name} ---')
    print(merged_data.head())
    print('\n' + '-' * 50 + '\n')

    # 결과 저장
    merged_data.to_csv(f'merged_data_{name}.csv')
```
