```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#  
model_data = pd.read_csv('datasets/ks200.csv', encoding='cp949')

#   . 'forward_stage' 'Date' .
features = model_data.drop(columns=['forward_stage', 'Date'])

#   .
correlation_matrix = features.corr().round(2)

#  .
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#  
model_data = pd.read_csv('datasets/ks200.csv', encoding='cp949')

#   . 'forward_stage' 'Date' .
features = model_data.drop(columns=['forward_stage', 'Date'])

#   .
correlation_matrix = features.corr()

# 0.7      .
high_corr_pairs = {}
for column in correlation_matrix.columns:
    for index in correlation_matrix.index:
        if 0.7 < abs(correlation_matrix.loc[index, column]) < 1:  #  0.7  1  
            if (column, index) not in high_corr_pairs and (index, column) not in high_corr_pairs:  #   
                #    
                high_corr_pairs[(index, column)] = round(correlation_matrix.loc[index, column], 2)

high_corr_pairs
```
